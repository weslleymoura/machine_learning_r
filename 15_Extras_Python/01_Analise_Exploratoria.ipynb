```python
# Importa as bibliotecas
import pandas
import matplotlib.pyplot as plt
import numpy 
from pandas.tools.plotting import scatter_matrix
```

```python
# Carrega os dados
url = "https://goo.gl/vhm1eU"
colunas = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataset = pandas.read_csv(url, names=colunas)
```

```python
# PANDAS: Verificando alguns dados
exemplos = dataset.head(10)
print(exemplos)
```

```python
# PANDAS: Verifica a dimensão dos dados (linhas, colunas)
dim = dataset.shape
print(dim)
```

```python
# PANDAS: Verifica os tipos de dados de cada atributo
tipos = dataset.dtypes
print(tipos)
```

```python
# PANDAS: Estatística descritiva
pandas.set_option('display.width', 100)
pandas.set_option('precision', 3)
resultado = dataset.describe()
print(resultado)
```

```python
# PANDAS: Distribuição das classses
contagem = dataset.groupby('class').size()
print(contagem)
```

```python
# PANDAS: Correlação
cor = dataset.corr(method='pearson')
print(cor)
```

```python
# PANDAS: Skew
skew = dataset.skew()
print(skew)
```

```python
# MATPLOTLIB: Histograma (univariado)
dataset.hist()
plt.show()
```

```python
# MATPLOTLIB: Gráfico de densidade (univariado)
dataset.plot(kind='density', subplots=True, layout=(3,3), sharex=False)
plt.show()
```

```python
# MATPLOTLIB: Boxplot e Whisker com uma variável
dataset.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
plt.show()
```

```python
# MATPLOTLIB + Numpy: Gráfico de correlação (multivariado)
cor = dataset.corr()

# Monta o gráfico
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cor, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(colunas)
ax.set_yticklabels(colunas)
plt.show()
```

```python
# Gráfico de dispersão (multivariado) 
scatter_matrix(dataset)
plt.show()
```
